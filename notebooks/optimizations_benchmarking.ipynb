```python
import pandas as pd
import matplotlib.pyplot as plt

inference = pd.read_csv("../src/multilora/benchmarking/results_inference.csv")
tuning = pd.read_csv("../src/multilora/benchmarking/results_tuning.csv")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(inference['n_adapters'], inference['no_stk'], marker='*', label='Grouping', )
plt.plot(inference['n_adapters'], inference['stk'], marker='*', label='Megablocks')
plt.xlabel("Number of adapters")
plt.ylabel("Thoughput (req/sec)")
plt.grid()
plt.legend()
plt.savefig("Inference_res.pdf", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(tuning['n_adapters'], tuning['no_stk'] * 16, marker='*', label='Grouping', )
plt.plot(tuning['n_adapters'], tuning['stk'] * 16, marker='*', label='Megablocks')
plt.xlabel("Number of adapters")
plt.ylabel("Thoughput (tok/sec)")
plt.grid()
plt.legend()
plt.savefig("finetuning_res.pdf", dpi=300)
plt.show()
```
